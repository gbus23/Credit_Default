```python
import requests
import pandas as pd
import requests
from requests_negotiate_sspi import HttpNegotiateAuth
pd.set_option('display.max_columns', None)
```

```python
def fetch_dataframe_from_URL(url):
    try:
        # Use HttpNegotiateAuth for Windows authentication
        response = requests.get(url, verify=False, auth=HttpNegotiateAuth())
        response.raise_for_status()
        data = response.json()
        return pd.DataFrame(data)
    except requests.exceptions.RequestException as e:
        print("Error fetching data:", e)
        return None
```

```python
url = "https://smartbase.tikehaucapital.com/API/CMS/Transactions?startTradeDate=2024-01-01"
df= fetch_dataframe_from_URL(url)
```
